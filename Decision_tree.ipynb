```python
import numpy as np 
import pandas as pd 
```

```python
df = pd.read_csv('covid_toy.csv')
```

```python
df.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
```

```python
si = SimpleImputer()
```

```python
df['fever'] = si.fit_transform(df[['fever']])
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb = LabelEncoder()
```

```python
df.head()
```

```python
df['gender'] = lb.fit_transform(df['gender'])
df['fever'] = lb.fit_transform(df['fever'])
df['city'] = lb.fit_transform(df['city'])
df['has_covid'] = lb.fit_transform(df['has_covid'])
```

```python
df.head()
```

```python
# drop the has covid column
x= df.drop(columns = ['has_covid'])
y = df['has_covid']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier()
```

```python
dt.fit(x_train, y_train)
```

```python
y_pred = dt.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_pred)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
# created a parameter grid based on the results of random search
param_grid = {
    'max_depth' : [2, 3, 5, 10, 20],
    'min_samples_leaf' : [5, 10, 20,50,100],
    'criterion' : ['gini', 'entropy'],
}
```

```python
grid_search = GridSearchCV(estimator=dt, param_grid=param_grid, cv=6, n_jobs=-1, verbose=2 , scoring='accuracy')
```

```python
grid_search.fit(x_train, y_train)
```

```python
grid_search.best_estimator_
```

```python
y_pred_2 = grid_search.predict(x_test)
```

```python
accuracy_score(y_test, y_pred_2)
```

```python
df = pd.read_csv('Social_Network_Ads.csv')
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb = LabelEncoder()
```

```python
df['Gender'] = lb.fit_transform(df['Gender'])
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
x = df.drop(columns = ['Purchased'])
y = df['Purchased']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier()
```

```python
dt.fit(x_train, y_train)
```

```python
y_pred = dt.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
# created a parameter grid based on the results of random search
param_grid = {
    'max_depth' : [2, 3, 5, 10, 20],
    'min_samples_leaf' : [5, 10, 20,50,100],
    'criterion' : ['gini', 'entropy'],
}
```

```python
grid_search = GridSearchCV(estimator=dt, param_grid=param_grid, cv=6, n_jobs=-1, verbose=2 , scoring='accuracy')
```

```python
grid_search.fit(x_train, y_train)
```

```python
grid_search.best_estimator_
```

```python
y_pred_2 = grid_search.predict(x_test)
```

```python
accuracy_score(y_test, y_pred_2)
```
