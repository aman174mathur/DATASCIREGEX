```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('titanic.csv' , usecols=['Survived' ,'Pclass' , 'Age' , 'Sex' , 'SibSp' , 'Parch' ,'Fare' , 'Embarked'])
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer

# Impute missing values in the 'Age' column with the mean
si = SimpleImputer()
df['Age'] = si.fit_transform(df[['Age']])
df['Fare'] = si.fit_transform(df[['Fare']])
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Sex'] = le.fit_transform(df['Sex'])
df['Age'] = le.fit_transform(df['Age'])
df['Embarked'] = le.fit_transform(df['Embarked'])
```

```python
df.head()
```

```python
x = df.drop(columns= ['Survived'])
y = df['Survived']
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x, y , test_size=0.2 , random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train_sc = sc.fit_transform(x_train)
x_test_sc = sc.transform(x_test)
```

```python
from sklearn.naive_bayes import  BernoulliNB
classifier = BernoulliNB()
classifier.fit(x_train_sc,y_train)
BernoulliNB()
```

```python
y_pred = classifier.predict(x_train_sc)
y_pred
```

```python
from sklearn.metrics import confusion_matrix

# Generate predictions for the test set
y_pred_test = classifier.predict(x_test_sc)

# Compute the confusion matrix
cn = confusion_matrix(y_test, y_pred_test)
```

```python
cn
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred_test)
```

```python
from sklearn.metrics import precision_score
precision_score(y_test,y_pred_test)
```

```python
from sklearn.metrics import recall_score
recall = recall_score(y_test , y_pred_test)
recall
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()

# Fit the model
lr.fit(x_train, y_train)

# Make predictions
y_pred_lr = lr.predict(x_test)
y_pred_lr
```
