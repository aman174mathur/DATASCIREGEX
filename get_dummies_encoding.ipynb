```python
# what does it get dummies do ?
# it creates a new column for each category 
# in the categorical variable
# and assigns a 1 or 0 to each row 
# based on whether the row belongs to that category or not

# it is used to convert categorical variables into numerical variables
# so that they can be used in machine learning models that require 
# numerical variables
```

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("covid_toy.csv")
```

```python
df.head()
```

```python
df.dropna(inplace= True)
```

```python
df.isna().sum()
```

```python
df_dummies=pd.get_dummies(df,columns=['gender','cough','city','has_covid'], drop_first=True)
df_dummies.astype(int)
```
